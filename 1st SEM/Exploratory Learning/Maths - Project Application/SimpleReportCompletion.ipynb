# Monte Carlo Simluation

```python
# Import NumPy
import numpy as np

# Conduct 1 million RANDOM simulations
simulations = 1000000

reportA = np.random.uniform(1, 5, simulations)
reportB = np.random.uniform(2, 6, simulations)

# duration, stores the million random samples
duration = reportA + reportB
```

### Visualization of Result

```python
# Import Matplotlib
import matplotlib.pyplot as plt

# Plot Histogram with 
# (approximate probability densities) and (line at value 9)
plt.figure()
plt.hist(duration, density = True, edgecolor='white')
plt.axvline(9, color='r')
plt.show()

# Percentage of simulations exceeding 9 hours
print((duration > 9).sum()/simulations)
```
